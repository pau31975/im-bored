```python
import random

def trade_simulate(trade_start, num_trade, percent_win, percent_save_win):

    # Set the percentage of True and False
    percent_win = percent_win
    percent_lose = 1 - percent_win

    # Generate the list
    win_lose_list = random.choices([True, False], weights=[percent_win, percent_lose], k=num_trade)
    # print(win_lose_list)

    trade_list = [(trade_start, 0)]

    for i in win_lose_list:

        diff = 0
        
        if i == True:
            diff = trade_list[-1][0] * random.uniform(0, 0.03)
            trade_list.append((trade_list[-1][0] + diff * (1 - percent_save_win), trade_list[-1][1] + diff * percent_save_win))
        else:
            diff = trade_list[-1][0] * random.uniform(-0.02, 0)
            trade_list.append((trade_list[-1][0] + diff, trade_list[-1][1]))
        
    # print(win_lose_list)
    return trade_list
```

```python
for i in trade_simulate(100, 10, 0.6, 0):
    print(i)
    print(i[0]+i[1])
```

```python
import matplotlib.pyplot as plt

values = trade_simulate(100, 30, 0.5)

plt.plot(values)
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Values in the List')
plt.plot(range(len(values)), [100] * len(values), 'r--', label=f'initial: {100}')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample list of values
values = []

for i in range(10):
    values.append(trade_simulate(100, 100, 0.6)[-1])

plt.bar(range(len(values)), values)
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Values in the List')
plt.plot(range(len(values)), [100] * len(values), 'r--', label=f'initial: {100}')
plt.legend()
plt.show()
```
